# MATPLOTLIB for data visulisation 

```python
import matplotlib.pyplot as plt

x = [2,5,8,9,12]
y = [1,2,3,4,5]
```

```python
x
```

```python
y
```

```python
#to plot
plt.plot(x,y)

#labels
plt.xlabel("years")
plt.ylabel("sales")

#title
plt.title("Sales vs. Year")
```

```python
#to plot
plt.plot(x,y,linestyle='dotted')

#labels
plt.xlabel("years")
plt.ylabel("sales")

#title
plt.title("Sales vs. Year")
```

```python
#to plot
plt.plot(x,y,color='red',linestyle='dotted')

#labels
plt.xlabel("years")
plt.ylabel("sales")

#title
plt.title("Sales vs. Year")
```

```python
#to plot
plt.plot(x,y,color='red',linestyle='dotted',marker='o')

#labels
plt.xlabel("years")
plt.ylabel("sales")

#title
plt.title("Sales vs. Year")
```

```python
#to plot
plt.plot(x,y,color='orange',linestyle='dotted',marker='*',ms=15)

#labels
plt.xlabel("years")
plt.ylabel("sales")

#title
plt.title("Sales vs. Year")
```

```python
#to plot
plt.bar(x,y,color='red')

#labels
plt.xlabel("years")
plt.ylabel("sales")

#title
plt.title("Sales vs. Year")
```

```python
#to plot
plt.scatter(x,y)
plt.bar(x,y)

#labels
plt.xlabel("years")
plt.ylabel("sales")

#title
plt.title("Sales vs. Year")
```

# subplot

```python
import numpy as np
```

```python
a = np.array([44,87,215,48,654])
b = np.array([4,857,5,484,56])
c = np.array([54,456,45,44,54])
d = np.array([54,84,62,78,13])
```

```python
plt.subplot(2,3,1)
plt.plot(a,b,color='red')

plt.subplot(2,3,2)
plt.plot(a,d,color='orange')

plt.subplot(2,3,3)
plt.plot(c,d,color='purple')

plt.subplot(2,2,3)
plt.plot(b,d,color='maroon')

plt.subplot(2,2,4)
plt.plot(d,c,color='pink')
```
